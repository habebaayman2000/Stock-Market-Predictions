```python
# LSTM

#Training & Testing

data_training=pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing=pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])
```

```python
data_training.shape
```

```python
data_testing.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))

data_training_array=scaler.fit_transform(data_training)
```

```python
data_training_array
```

```python
data_training_array.shape[0]
```

```python
data_training_array[51,0]
```

```python
x_train=[]
y_train=[]

for i in range(50,data_training_array.shape[0]):
    x_train.append(data_training_array[i-50:i])
    y_train.append(data_training_array[i,0])


x_train,y_train= np.array(x_train),np.array(y_train)
```

```python
x_train.shape
```

```python
# Model Bulding

from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
# LSTM Input -> 3D Array (batch_size, time_steps, seq_len)
# LSTM Output 2D OR 3d
# 2D -> (batch_size, units)
# 3D -> (batch_size, time_steps, units)
```

```python
x_train.shape[1]
```

```python
model=Sequential()
model.add(LSTM(units=50,activation='relu',return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=60,activation='relu',return_sequences=True))
model.add(Dropout(0.3))

model.add(LSTM(units=80,activation='relu',return_sequences=True))
model.add(Dropout(0.4))

model.add(LSTM(units=120,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1 ))

model
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(x_train,y_train,epochs=50)
```

```python
def plot_graph(figsize,values,colunm_names):
  plt.figure(figsize=figsize)
  values.plot()
  plt.xlabel('years')
  plt.ylabel(colunm_names)
  plt.title(f'{colunm_names} over years')
  plt.legend()
  plt.grid(True)
  plt.show()
```

```python
google_data['pct_change']=google_data['Close'].pct_change()
google_data['pct_change']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
scaler_data=scaler.fit_transform(google_data['Close'])
```

```python
scaler_data
```

```python
len(scaler_data)
```

```python
x_data=[]
y_data=[]

for i in range(100,len(scaler_data)):
    x_data.append(scaler_data[i-100:i])
    y_data.append(scaler_data[i])


x_data,y_data= np.array(x_data),np.array(y_data)
```

```python
splitting_data=int(len(x_data)*0.70)
x_train=x_data[:splitting_data]
y_train=y_data[:splitting_data]

x_test=x_data[splitting_data:]
y_test=y_data[splitting_data:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Model Bulding

from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model=Sequential()
model.add(LSTM(128,return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(64,return_sequences=False))
model.add(Dense(25))
model.add(Dense(units=1 ))

model
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(x_train,y_train,epochs=2,batch_size=1)
```

```python
prediction=model.predict(x_test)
```

```python
prediction
```

```python
inverse_prediction=scaler.inverse_transform(prediction)
inverse_prediction
```

```python
inverse_ytest=scaler.inverse_transform(y_test)
inverse_ytest
```

```python
rmse=np.sqrt(np.mean(inverse_prediction-inverse_ytest)**2)
```

```python
rmse
```

```python
plotting_data=pd.DataFrame(
    {
        'Actual':inverse_ytest.reshape(-1),
        'Predicted':inverse_prediction.reshape(-1)
    },index= google_data.index[splitting_data+100:]
)
plotting_data.head()
```

```python
plot_graph((15,5),plotting_data,'Predicted vs Actual')
```

```python
plot_graph((15,5),pd.concat([google_data['Close'][:splitting_data+100],plotting_data]),'Predicted vs Actual')
```
